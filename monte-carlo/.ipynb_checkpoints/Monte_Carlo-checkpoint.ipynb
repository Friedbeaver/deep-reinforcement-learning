### Part 1: MC Prediction

We will begin by investigating a policy where the player _almost_ always sticks if the sum of her cards exceeds 18.  In particular, she selects action `STICK` with 80% probability if the sum is greater than 18; and, if the sum is 18 or below, she selects action `HIT` with 80% probability.  The function `generate_episode_from_limit_stochastic` samples an episode using this policy. 

The function accepts as **input**:
- `bj_env`: This is an instance of OpenAI Gym's Blackjack environment.

It returns as **output**:
- `episode`: This is a list of (state, action, reward) tuples (of tuples) and corresponds to $(S_0, A_0, R_1, \ldots, S_{T-1}, A_{T-1}, R_{T})$, where $T$ is the final time step.  In particular, `episode[i]` returns $(S_i, A_i, R_{i+1})$, and `episode[i][0]`, `episode[i][1]`, and `episode[i][2]` return $S_i$, $A_i$, and $R_{i+1}$, respectively.

````python
def generate_episode_from_limit_stochastic(bj_env):
    episode = []
    state = bj_env.reset()
    while True:
        probs = [0.8, 0.2] if state[0] > 18 else [0.2, 0.8]
        action = np.random.choice(np.arange(2), p=probs)
        next_state, reward, done, info = bj_env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
    return episode
````

Play Blackjack with the policy from above.

````python
for i in range(3):
    print(generate_episode_from_limit_stochastic(env))
````

**Input**
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `generate_episode`: This is a function that returns an episode of interaction.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

**Output**
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

**Note:** Many seemingly disparate algorithms in Reinforcement Learning can be unified.
- The on-policy variation of the algorithm is recovered from the off-policy case, if the behavior $b$ and target $\pi$ policy are equal (and the importance sampling ratio $W$ is 1).

````python
def mc_prediction_q(env, num_episodes, generate_episode, gamma=1.0):
    N = defaultdict(lambda: np.zeros(env.action_space.n))
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        # generate episode 
        episode = generate_episode_from_limit_stochastic(env)
        G = 0 # reset initial return to zero prior to starting any episode
        # go through each timestep of the episode
        for (state, action, reward) in reversed(episode):
            G = gamma*G + reward # accumulate sum of discounted rewards
            # count every-visit to state-action pair
            N[state][action] += 1
            # average returns incrementally
            Q[state][action] += 1/N[state][action]*(G - Q[state][action])
    
    return Q
````

Obtain the action-value function estimate $Q$.
Plot the corresponding state-value function.

````python
# obtain the action-value function
Q = mc_prediction_q(env, 500000, generate_episode_from_limit_stochastic)

# obtain the corresponding state-value function
V_to_plot = dict((k,(k[0]>18)*(np.dot([0.8, 0.2],v)) + (k[0]<=18)*(np.dot([0.2, 0.8],v))) \
         for k, v in Q.items())

# plot the state-value function
plot_blackjack_values(V_to_plot, './results/every_visit_mc_prediction.png')
````

### Part 2: On-Policy MC Control (for $\epsilon$-soft policies)

**Input**
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

**Output**
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.
- `policy`: This is a dictionary where `policy[s]` returns the action that the agent chooses after observing state `s`.

**Note:** Many seemingly disparate algorithms in Reinforcement Learning can be unified.
- If $\alpha$ is set to the fraction of counts of visits to node $s$, we have a non-constant learning step.
- The on-policy variation of the algorithm is recovered from the off-policy case, if the behavior $b$ and target $\pi$ policy are equal (and the importance sampling ratio $W$ is 1).

Get the $\epsilon$-soft probabilities.

````python
def get_probs(Q_s, nA, epsilon):
    probs = np.ones(nA)*epsilon/nA
    best_action = np.argmax(Q_s)
    probs[best_action] += 1 - epsilon
    
    return probs  
````

Get an episode of (state, action, reward) tuples behaving according to an $\epsilon$-greedy policy.

````python
def generate_episode_from_eps_greedy_policy(env, Q, epsilon):
    episode = []
    state = env.reset()
    nA = env.action_space.n
    while True:
        # one-step transition
        probs = get_probs(Q[state], nA, epsilon)
        # sample action using epsilon-greedy policy
        action = np.random.choice(np.arange(nA), p=probs)
        next_state, reward, done, info = env.step(action)
        episode.append((state, action, reward))
        state = next_state
        if done:
            break
            
    return episode
````

````python
def mc_control(env, num_episodes, alpha, gamma=1.0, eps_min=0.05, eps_start=1.):
    nA = env.action_space.n
    N = defaultdict(lambda: np.zeros(nA))
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    policy = defaultdict(lambda: np.zeros(nA))
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        # decay epsilon appropiately
        epsilon = eps_start/i_episode
        epsilon = max(epsilon, eps_min)
        
        # generate episode
        episode = generate_episode_from_eps_greedy_policy(env, Q, epsilon)
        G = 0 # reset initial return to zero prior to starting any episode
        # go through each timestep of the episode
        for (state, action, reward) in reversed(episode):
            G = gamma*G + reward # accumulate sum of discounted rewards
            # count every-visit to state-action pair
            N[state][action] += 1
            # average returns incrementally
            if alpha is None:
                alpha = 1/N[state][action]
            Q[state][action] += alpha*(G - Q[state][action])
            probs = get_probs(Q[state], nA, epsilon)
            for a in range(nA):
                policy[state][a] = np.random.choice(np.arange(nA), p=probs)
        
    return policy, Q
````

Obtain the estimated optimal policy and action-value function.

````python
# obtain the estimated optimal policy and action-value function
policy, Q = mc_control(env, 500000, 0.02)
````

Next, we plot the corresponding state-value function.

````python
# obtain the corresponding state-value function
V = dict((k,np.max(v)) for k, v in Q.items())

# plot the state-value function
plot_blackjack_values(V, './results/on_policy_every_visit_mc_control_const_alpha.png')
````

Finally, we visualize the policy that is estimated to be optimal.

````python
# plot the policy
policy = dict((k,np.argmax(v)) for k, v in Q.items()) # make it deterministic
plot_policy(policy, './results/policy_of_on_policy_every_visit_mc_control_const_alpha.png')
````
